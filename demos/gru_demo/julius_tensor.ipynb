## Julius Tensor 

In this demo we evaluate pre-trained GRUs trained on the complete works of 
William Shakespeare.



```python
import tensorflow as tf
import logging
tf.get_logger().setLevel(logging.ERROR)
```

```python
def generate_shakespeare(model, size=1000, seed='ROMEO:'):
    states = None
    next_char = tf.constant([seed])
    result = [next_char]
                             
    for n in range(size):
      next_char, states = model.generate_one_step(next_char, states=states)
      result.append(next_char)

    return tf.strings.join(result)[0].numpy().decode("utf-8")
                             
```

###  Word play

We have upload three pre-trained models for you to experiment with: 'Hamlet', 'Macbeth', and 'Comedy of Errors'. One of the models has 256 hidden neurons in the GRU layer, and was trained
for 30 epochs. The two other models each have have 1024 hidden neurons in the GRU layer; one was trained for 30 epochs, and the other trained for 50 epochs. 

Generate sample text from each, using the function `generate_shakespeare` defined above, in order to evaluate the models. Can you tell which model is which? 


```python
pretrained1 = tf.saved_model.load('Comedy_of_Errors')
pretrained2 = tf.saved_model.load('Macbeth')
pretrained3 = tf.saved_model.load('Hamlet')
```

```python
print(generate_shakespeare(pretrained1, size=10000, seed='ROMEO:'))
```

```python
print(generate_shakespeare(pretrained2, size=10000, seed='ROMEO:'))
```

```python
print(generate_shakespeare(pretrained3, size=10000, seed='ROMEO:'))
```
